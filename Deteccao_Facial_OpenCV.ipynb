```python

# Deteco Facial com OpenCV (Haar Cascades)
# Autores: Luiz Fernando e Ismael
# Descrio: Este notebook detecta rostos humanos em imagens usando classificadores Haar.

import cv2
import os
import matplotlib.pyplot as plt

# Caminho base com as imagens organizadas por pessoa
base_path = "imagens"

# Carregar o classificador Haar Cascade para deteco facial
cascade_path = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
face_cascade = cv2.CascadeClassifier(cascade_path)

# Funo para mostrar imagem com rostos detectados
def detectar_rostos_e_exibir(image_path):
    imagem = cv2.imread(image_path)
    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

    rostos = face_cascade.detectMultiScale(imagem_cinza, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in rostos:
        cv2.rectangle(imagem, (x, y), (x+w, y+h), (0, 255, 0), 2)

    imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
    plt.imshow(imagem_rgb)
    plt.title(f"Rostos detectados: {len(rostos)}")
    plt.axis("off")
    plt.show()

# Executar deteco para a 5 imagem de cada pessoa
for pessoa in os.listdir(base_path):
    person_folder = os.path.join(base_path, pessoa)
    for img_name in os.listdir(person_folder):
        if "img_05" in img_name:
            img_path = os.path.join(person_folder, img_name)
            print(f"Analisando: {img_path}")
            detectar_rostos_e_exibir(img_path)

```
